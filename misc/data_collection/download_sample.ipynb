```python
import talent_list
names = talent_list.name_list
```

```python
import os
import requests
import time

# Create the directory if it doesn't exist
os.makedirs("dist", exist_ok=True)

# Base URL
base_url = "https://samplecontents-dcon.groovy-ec.com/public/shop-nijisanji-niji/import/assets/"

# Download each file
for name in names:
    file_url = f"{base_url}{name} 誕生日ボイス2024 サンプル.mp3"
    response = requests.get(file_url)
    
    if response.status_code == 200:
        with open(f"sample-voice/{name}_誕生日ボイス2024_サンプル.mp3", "wb") as file:
            file.write(response.content)
        print(f"Downloaded: {name}_誕生日ボイス2024_サンプル.mp3")
    else:
        print(f"Failed to download {name}'s file. Status code: {response.status_code}")

    time.sleep(1)
```
